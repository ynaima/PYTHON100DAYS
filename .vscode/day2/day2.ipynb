Today's errors
1. Type error - using a function with data of the wrong type
2. Value error - converting object to an object that it cannot possibly be converted to.

```python
#a)Strings - characters - anything wrapped in quotes
#lets start with subscripting or string slicing
print("Hero"[2])
```

1. From the beginning, indexing starts with 0
2. from the last character, it starts with indexing -1
3. If we want to print characters in reverse we use the negative index starting from the end of the string

```python
#last character
print("Hello"[-1])
#first character
print("Hello"[0])
#second character from the end
print("Hello"[-2])
```

```python
#String concatenation with numbers
print("123" + "345")
```

```python
#b) Integers - whole numbers
#this will do an actual mathematical addition because the data type used here is integers and not strings
print(123 + 345)
```

TypeErros - these are errors that occur when we use a function that is associated with a certain data type with data that is not of that data type
example as follows.

```python
#In this case the len  function is associated with strings but we are using it with integers
print(len(45678))
```

```python
#we can fix this code by changing the data to a string
print(len("45678"))
```

Type() function - returns the data type of an object

```python
#challenge1: print 4 examples of the data types we learned and their data type checks using the type function
print(type(4567))
print(type("hello"))
print(type(456.78))
print(type(False))
```

Type Conversion / Type parsing
Converting objects of 1 type to another (only valid ones)
bool()
float()
str()
int()

```python
#converting string object to integer
print(int("123") + int("456"))
```

```python
#converting it to strings
print(str(123) + str(456))
```

Mathematical operations in python
** used as powers
// integer division - returns whole number 
/ returns floats
% module operator returns reminder

Operator Precedence - from left to right
PEMDAS
()
**
* OR /
+ OR -

```python
print(2**3)
print(5/2)
print(5//2)
print(5%2)
print(456+657)
print(464-221)

print(3*3 + 9 - 3/3 + (2**3) + (4*2))
```

```python
#round function
#if we want intger we just use round without specifying decimal number details
print(round(45.6778999))

print(round(45.3778999))
```

```python
#if we want float number value
print(round(45.6778999, 3))
```

Assignment Operator
we can use any operator in place of the + symbol
score +=2 is the same as score = score + 2

```python
score = 0
score +=2
print(score)
```

f strings allow you to incorporate different data types into 1 print statement without manually doing type conversion

```python
score = 10
height = 1.65
is_winning = True

print(f"Your score is = {score}, your height is {height}. Your winning is {is_winning}.")
```
