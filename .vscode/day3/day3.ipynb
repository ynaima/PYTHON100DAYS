#challenge2
Python Pizza Delivery Program
Introduction to the Python Pizza Delivery Program
In this lesson, we have a small coding challenge: to create a Python Pizza Delivery Program. This program will automatically calculate the bill for the user based on several options. The user will be asked three questions through input prompts.

When running the program, it first displays "Welcome to Python Pizza Deliveries!" and then asks, "What size pizza do you want?" The user can choose from S, M, or L. For example, choosing L for large.

Next, it asks, "Do you want pepperoni on your pizza?" and then "Do you want extra cheese?" The answers are saved into variables: size, pepperoni, and extra_cheese.

The program will then print out the final bill amount, such as "Your final bill is..." followed by the calculated price. For instance, the final bill could be $28 based on the pricing rules.

Pricing Details
Small pizza costs $15
Medium pizza costs $20
Large pizza costs $25
Additional charges:

Pepperoni on a small pizza: +$2
Pepperoni on medium or large pizza: +$3
Extra cheese: +$1

```python
print("Welcome to Python Pizza Deliveries!")

size = input("What size pizza do you want? Type S for small, M for medium and L for large\n")

pepperoni = input("Do you want pepperoni on your pizza? Type y for yes, n for no?\n")

extra_cheese = input("Do you want extra cheese on your pizza? Type y for yes, n for no?\n")

bill = 0

#checking the size
if size == "S":
    print("Small pizza costs $15")
    bill += 15

elif size == "M":
    print("Medium pizza costs $20")
    bill += 20

elif size == "L":
    print("Large pizza costs $25")
    bill += 25

else:
    print("Wrong input, please try again!")


#checking the toppings
if pepperoni == "y":
    print("pepperoni is extra $2 for small size pizza and $3 for medium or large")
    if size == "S":
        bill += 2
    else:
        bill +=3


if extra_cheese == "y":
    print("extra cheese is $1 on any pizza size")
    bill += 1


print(f"Your final bill is ${bill}.")



```

Logical Operators
We can use logical operators with multiple conditions
if condition1 & condition2 & condition3:
    do this
else:
    do this

we have the following logical operators
AND
A and B  (both have to be True for the condition to be True)
OR   
A or B  (either A or B have to be True or both)

NOT 
not A (reverses the condition - if it was True it becomes False and if it was false it becomes True)
