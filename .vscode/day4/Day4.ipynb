List order and indexing
Just like string, list items can be accessed by their indexing marking their position in the list 
First item is 0 last item is -1 or len(list)-1 in that order
The first item is 0 shows the offset from the start is 0 while the second item shows offset from the begnning is 1

```python
my_list = ["apple", "cherry", 56, "123", 34]

print(f"first item {my_list[0]}")

print(f"last item {my_list[-1]}")
 
print(f"or {my_list[len(my_list)-1]}")

print(f"third item {my_list[2 ]}")

print(f"second item from the end of the list {my_list[-2]}")

print(f"start stop step - display every second item {my_list[::2]}")

print(f"start stop step - display every second item from the end{my_list[::-2]}")

print(f"reverse the list {my_list[::-1]}")

print(f"display the whole list {my_list[::]}")
```

Nested Lists - a list within another list

```python
fruits = ["Strawberries", "Nectarines", "Apples", "Grapes", "Peaches", "Cherries", "Pears"]
vegetables = ["Spinach", "Kale", "Tomatoes", "Celery", "Potatoes"]
dirty_dozen = [fruits, vegetables]
dirty_dozen
```

```python
#to access first item in the second list items in the list 
#first square bracket - position of the list in the big list
#second square bracket -position of the item in the identified list 
print(f"first item in the second list {dirty_dozen[1][0]}")
```
